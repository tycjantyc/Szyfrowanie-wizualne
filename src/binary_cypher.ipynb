```python
import numpy as np
from skimage import io
from skimage import color
import matplotlib.pyplot as plt
import random


image = io.imread("../resources/mountain.jpg")
image = color.rgb2gray(image)

print(image.shape)
Y,X = image.shape
obraz_1 = np.zeros((2*Y,2*X))
obraz_2 = np.zeros((2*Y,2*X))

for y in range(Y):
    for x in range(X):
        if image[y,x] >= 0.5:
            image[y,x] = 1
        else:
            image[y,x] = 0


plt.imshow(image)
count=0
for y in range(Y):
    for x in range(X):
        
        temp = np.random.randint(2, size=(2,2))
        temp_2 = -(temp- np.ones((2,2)))
        obraz_1[2*y:2*(y+1), 2*x:2*(x+1)] = temp
        
        if image[y,x]:
            obraz_2[2*y:2*(y+1), 2*x:2*(x+1)] = temp_2
        else:
            obraz_2[2*y:2*(y+1), 2*x:2*(x+1)] = temp



np.array_equal(obraz_1, obraz_2)
```
